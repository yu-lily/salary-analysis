```python
df = df[df['location'] == 'San Francisco, CA']
```

```python
pt = df.pivot_table(index='company', columns='gender', aggfunc='count', values='location')#.sort_values('Non-binary', ascending=False)[:10]
pt['total'] = pt.sum(axis=1)
pt['gay_pct'] = pt['Non-binary'] / pt['total']
pt['pct_techbros'] = pt['Male'] / pt['total']
```
