```python
import pandas as pd
import json
import numpy as np
from collections import defaultdict
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
df = pd.read_csv('data/processed.tsv', sep='\t')
```

```python
df2 = pd.read_csv('data/pm-processed.tsv', sep='\t')
```

```python
df = df[df['yoe_total'] <= 35]
df2 = df2[df2['yoe_total'] <= 35]
df = df[df['degree'] != 'phd']
df2 = df2[df2['degree'] != 'phd']
```

```python
sns.lineplot(data=df, x='yoe_total', y='tc', label='SWE')
sns.lineplot(data=df2, x='yoe_total', y='tc', label='PM')
plt.savefig('./vis/pm-vs-swe.png')
```

```python
df.groupby('yoe_total').mean()['tc']
```

```python
FAANG = ['Facebook', 'Amazon', 'Apple', 'Netflix', 'Microsoft', 'Google']
df = df[df['company'].isin(FAANG)]
df2 = df2[df2['company'].isin(FAANG)]
```

```python
sns.lineplot(data=df, x='yoe_total', y='tc', label='SWE')
sns.lineplot(data=df2, x='yoe_total', y='tc', label='PM')
plt.savefig('./vis/pm-vs-swe-faang.png')
```

```python
sns.histplot(data=df, x='yoe_total', label='SWE')
sns.histplot(data=df2, x='yoe_total', label='PM')
```
