```python
import pandas as pd
```

```python
df = pd.read_csv('data/main-processed.tsv', sep='\t')
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'] < '2020-02-01']
```

```python
df = df[['company', 'location', 'tag', 'gender', 'yoe_total', 'tc', 'base']]
df = df.dropna()
```

```python
top_20_company = df['company'].value_counts()[:15].index
df = df[df['company'].isin(top_20_company)]
```

```python
df = df[df['gender'].isin(['Male', 'Female'])]
```

```python
lyft_high = df[(df['company'] == 'Lyft') & (df['tc'] > 500000)].index[0]
df = df.drop(lyft_high)
```

```python
import os
import json
```

```python
if os.path.exists('./data/locations.json'):
    with open('./data/locations.json', 'r') as f:
        location_dict = json.load(f)
```

```python
location_dict
```

```python
def get_country(x):
    if x and location_dict[x]:
        return location_dict[x]['address']['country']
    return None
df['country'] = df['location'].apply(get_country)

def get_state(x):
    if x and location_dict[x]:
        if 'state' in location_dict[x]['address'].keys():
            return location_dict[x]['address']['state']
    return None
df['state'] = df['location'].apply(get_state)
```
