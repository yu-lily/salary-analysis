```python
import json
```

```python
with open('./data/sitemap.json', 'r', encoding='utf-8-sig') as f:
    sitemap = json.load(f)
    # data = f.read()
    # sitemap = json.loads(data)
```

```python
urls = []
for item in sitemap['urlset']['url']:
    url = item['loc'].split('//')[-1]
    
    urls.append(url)
```

```python
urls
```

```python
from collections import defaultdict

def rec_dd():
    return defaultdict(rec_dd)
```

```python
url.split('/')
```

```python
urls_dict = rec_dd()
for url in urls:
    split_url = url.split('/')
    cur_dict = urls_dict
    
    for section in split_url[:-1]:
        if section:
            cur_dict = cur_dict[section]
    
    cur_dict[split_url[-1]] = url
```
